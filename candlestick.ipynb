```python
# a program to create a candle stick chart with plotly and python
import plotly.graph_objects as go 
import pandas as pd 
```

```python
# store data in var
df = pd.read_csv('Data/TSLA.csv')

# set indexes as date
df = df.set_index(pd.DatetimeIndex(df['Date'].values))

df.head(5)
```

```python
# create an interactive candlestick chart
figure = go.Figure(
    data = [
        go.Candlestick(
            x = df.index, 
            low = df['Low'],
            high = df['High'],
            close = df['Close'],
            open = df['Open'],
            increasing_line_color = 'green',
            decreasing_line_color = 'red'
        )
    ]
)

# make the horizontal zoom slider disappear
#figure.update_layout(xaxis_rangeslider_visible = False)

# add in titles
figure.update_layout(
    title = 'Tesla Price'
)

figure.update_xaxes(
    title_text = 'Date'
)

figure.update_yaxes(
    title_text = 'Tesla Stock Price USD ($)'
)

figure.show()
```
